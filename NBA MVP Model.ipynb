```python
#Creating List of each Link for MVP seasons (1980-2023)


roty = []


for x in range(1980,2024):

    link = f'https://www.basketball-reference.com/awards/awards_{x}.html'
    roty.append(link)
    
for x in range(0,len(roty)):

    #Turning MVP Files to CSVs for x in range(30,43):
    df = pd.read_html(roty[x],match = "Rookie of the Year")[0]
    df = df.droplevel(0, axis=1)
    df = df.dropna(axis = 0, how = 'all')


    df['ROTY Winner'] = np.round(df['Share'].nlargest (1))

    df['ROTY Winner'] = df['ROTY Winner'].fillna(0)

    df['Season'] = roty[x].split("_")[1].split(".")[0]


    df['Team'] = df['Tm'].map(team_dict)

    df['ID'] = df['Season'] + ' ' + df['Player'] + ' ' + df['Team']

    df['Team ID'] = df['Season'] + ' ' + df['Team']

    df.to_csv(f'C:\\Users\\Owner\\Documents\\Python Scripts\\NBA MVP Model files\\ROTYs\\{x+1980} Season ROTY.csv')
    
    time.sleep(10)
```

```python
#Creating List of each link for Advanced stats per season(1980-2024)

adv = []


for x in range(1980,2025):

    link = f'https://www.basketball-reference.com/leagues/NBA_{x}_advanced.html'
    adv.append(link)
    
for x in range(0,len(adv)):
    df = pd.read_html(adv[x], match='Advanced')[0]
    drop_RK = df[df['Rk'] == 'Rk']
    df = df.drop(drop_RK.index)
    advanced_list = ['Age', 'G', 'MP', 'PER', 'TS%', '3PAr',
           'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%',
           'OWS', 'DWS', 'WS', 'WS/48', 'OBPM', 'DBPM', 'BPM', 'VORP']
    df[advanced_list] = df[advanced_list].astype('float64')
    df = df.set_index('Rk')
#     df= df.drop(df[df['Tm'] == 'TOT'].index, axis=0)
    df['Team'] = df['Tm'].map(team_dict)
    df['Player'] = df['Player'].map(lambda title:title.rstrip('*'))
    df= df.drop(df[['Unnamed: 19', 'Unnamed: 24']], axis=1)
    df['Season'] = adv[x].split('_')[1][0:4]
    df['ID'] = df['Season'] + ' ' + df['Player'] + ' ' + df['Team']
    df['Team ID'] = df['Season'] + ' ' + df['Team']

    df.to_csv(f'C:\\Users\\Owner\\Documents\\Python Scripts\\NBA MVP Model files\\Advanced stats\\{x+1980} Advanced Stats.csv')
    
    time.sleep(10)
```

```python
#Creating List of each link for Advanced stats per season(1980-2024)

per = []


for x in range(1980,2025):

    link = f'https://www.basketball-reference.com/leagues/NBA_{x}_per_game.html'
    per.append(link)
    
for x in range(0,len(per)):
    df = pd.read_html(per[x], match='Player Per Game')[0]
    df = df.fillna(0)
    drop_RK2 = df[df['Rk'] == 'Rk']
    df = df.drop(drop_RK2.index)
    per_cols = ['Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%',
       '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']

    df[per_cols] = df[per_cols].astype('float64')
    df = df.set_index('Rk')
#     df= df.drop(df[df['Tm'] == 'TOT'].index, axis=0)
    df['Player'] = df['Player'].map(lambda title:title.rstrip('*'))
    df['Season'] = per[x].split('_')[1][0:4]
    df['GameScore'] = df['PTS'] + 0.4 * df['FG'] - 0.7 * df["FGA"] - 0.4*(df['FTA'] - df["FT"]) + 0.7 * df["ORB"] + 0.3 * df["DRB"] + df["STL"] + 0.7 * df["AST"] + 0.7 * df["BLK"] - 0.4 * df["PF"] - df["TOV"]
    df['Team'] = df['Tm'].map(team_dict)
    df['ID'] = df['Season'] + ' ' + df['Player'] + ' ' + df['Team']
    df['Team ID'] = df['Season'] + ' ' + df['Team']

    
    df.to_csv(f'C:\\Users\\Owner\\Documents\\Python Scripts\\NBA MVP Model files\\Player Per Game\\{x+1980} Per Game Stats.csv')
   
    
    time.sleep(7)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from itertools import combinations

#Removing current player stats
df = all_per_adv_team_norm_mvps_dpoy
df = df.drop(curr_per_adv_team_norm_mvps_dpoy.index)

df = df.fillna(df.mean())

columns = ['FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS','AST%', 'STL%', 'BLK%', 'TOV%', 'USG%', 'WS', 'BPM', 'VORP', 'SOS', 'SRS', 'Team ORtg', 'Team DRtg',
       'Team NRtg',  'Team TS%',"Rookie"]


# Assuming you have a DataFrame called 'df' with your data

# Create a list to store column names and their corresponding R-squared values
column_r2_scores = []

# Loop through all combinations of 33 columns
for combo in combinations(columns, 33):
    # Fit a linear regression model
    X = df[list(combo)]
    y = df['MVP Winner']  # replace 'target_variable' with your actual target variable
    model = LinearRegression().fit(X, y)
    
    # Predict the target variable
    y_pred = model.predict(X)
    
    # Calculate R-squared
    r2 = r2_score(y, y_pred)
    
    # Store column names and their R-squared values
    column_r2_scores.append((list(combo), r2))

# Sort the list by R-squared values in descending order
column_r2_scores.sort(key=lambda x: x[1], reverse=True)

# Get the top 10 columns with the highest R-squared values
top_10_columns = column_r2_scores[:10]

print("Top 10 columns with the highest R-squared values:")
for columns, r2 in top_10_columns:
    print(f"{columns}: {r2}")
```
